```python
code_writer= {"jhon","chriss","adam"}
print(code_writer)
drawer= {"lisa","beth","jhon"}
print(drawer)
print(code_writer.intersection(drawer))#hem kod yazan hemde çizim yapan
print(code_writer.union(drawer))#hepsi toplanacak ama jhonu sadece bir kere alıcak
print(code_writer.difference(drawer))#sadece kod yazan
print(drawer.difference(code_writer))#sadece çizim yapan
print(drawer.symmetric_difference(code_writer))#hem kod yazan hem çizim yapan hariç
```

```python
given_number = input("please, give me your number:")
given_number2 = input("please, give me your grade:")

given_number = int(given_number)
given_number2 = int(given_number2)

if given_number>5 and given_number2>50 :
  print("greater than 5 and greater than 50")
elif given_number<5 or given_number2<50 :
  print("less than 5 and less than 50")
else:
  print("equal to 5 and equal to 50")
```
